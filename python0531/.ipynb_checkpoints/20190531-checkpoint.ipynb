# 批量画图

```python
import os
```

```python
# load 数据
data = pd.read_excel('./testdata.xlsx')
```

```python
# 创建文件夹
for name in list(data['q_3'].unique()):
    os.makedirs('./output_data/%s'%name)

for name in list(data['q_3'].unique()):
    os.makedirs('./output_picture/%s'%name)
```

```python
def draw_map(data,college,name):
    data.plot(kind='bar')
    plt.savefig('./output_picture/%s/%s.png'%(college,name))
```

```python
# data_all 是第一部分数据
data_all = data.loc[:,'id':'q_3_5']
data_all = data_all.fillna(-1)
data_all = data_all.astype(int)
```

```python
def process(data_college,college):
    # 将所有字段整合到一起
    data_result = pd.DataFrame()
    data_percent = pd.DataFrame()
    for i in list(data_college.columns)[2:]:
        data = pd.DataFrame(data_college.loc[:,i].value_counts()).T
        percent = data/len(data_college)
        draw_map(data,college,i)
        data_result = data_result.append(data)
        data_percent = data_percent.append(percent)
    data_result.to_csv('./output_data/%s/data_result%s.csv'%(college,college))
    data_percent.to_csv('./output_data/%s/data_percent%s.csv'%(college,college))
```

```python
collegelist = list(data['q_3'].unique())
for college in collegelist:
    data_college = data_all[data_all['q_3']==college]
    process(data_college,college)
    print(college)
```

# 实际例子

### 预测泰坦尼克号生还率

```python
train = pd.read_csv('./titanic/train.csv')
test = pd.read_csv('./titanic/test.csv')
train[:4]
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
# 补全数据的方法有很多，这里利用中位数进行补全。
impute_value = train['Age'].median()
train['Age'] = train['Age'].fillna(impute_value)
test['Age'] = test['Age'].fillna(impute_value)
```

```python
# 对性别进行编码
train['IsFemale'] = (train['Sex'] == 'female').astype(int)
test['IsFemale'] = (test['Sex'] == 'female').astype(int)
```

```python
# 利用下面三个变量进行分析
predictors = ['Pclass', 'IsFemale', 'Age']
X_train = train[predictors].values
X_test = test[predictors].values
y_train = train['Survived'].values
X_train[:5]
```

```python
y_train[:5]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict[:10]
```

```python
# 这里查看准确率
model.score(X_train, y_train)
```
